```python
# Q) 1. The following data gives the geographical area (in acres) under paddy for 58 villages. Draw
# sample of eight villages by using SRSWOR. Based on sample drawn find the estimate of
# average area per village under paddy and estimate of its variance and 95% confidence
# interval for average area per village under paddy.
# 98, 19, 47, 137, 155, 270, 64, 69, 127, 292, 79, 81, 144, 104, 240, 273, 141, 56, 117, 201, 130, 58,
# 102, 170, 261, 158, 29, 102, 210, 189, 116, 46, 187, 101, 194, 93, 161, 222, 41, 127, 179, 223, 33,
# 114, 76, 96, 78, 88, 137, 114, 56, 108, 179, 318, 58, 58, 76, 272 
```

```python
data=np.array([98, 19, 47, 137, 155, 270, 64, 69, 127, 292, 79,81, 144,
               104, 240, 273, 141, 56, 117, 201, 130, 58,102, 170, 261,
               158, 29, 102, 210, 189, 116, 46, 187, 101, 194, 93,161,
               222, 41, 127, 179, 223, 33,114, 76, 96, 78, 88, 137, 114,
               56, 108, 179, 318, 58, 58, 76, 272])
```

```python
import scipy.stats
```

```python
y_bar=np.mean(data)       # Population mean
N=len(data)
n=8
s1=np.random.choice(data,n,replace=False)
s1
sy2=np.var(data)
# Estimate of the population mean Using sample(Average area per village Paddy)
Y_bar=np.mean(s1)
var_Y_bar=((N-n)/(n*N))*sy2    #SRSWOR  (Use population variance)
SE_Y_bar=np.sqrt(var_Y_bar)
t=scipy.stats.t.ppf(1-0.05/2,n-1)
LL=Y_bar-t*SE_Y_bar
UL=Y_bar+t*SE_Y_bar
df=pd.DataFrame([LL,UL,y_bar]).T
print(df.rename(columns={0:'LL',1:'UL',2:'y_bar'}))
# Estimated Variance of Y_bar and Estimated Standard Error of Y_bar
SY2=np.var(s1)
Est_var_Y_bar=((N-n)/(n*N))*SY2     
Est_SE_Y_bar=np.sqrt(Est_var_Y_bar)
# Estimated confidence Interval for y_bar
LL=Y_bar-t*Est_SE_Y_bar
UL=Y_bar+t*Est_SE_Y_bar
df1=pd.DataFrame([LL,UL,y_bar]).T
print(df1.rename(columns={0:'LL',1:'UL',2:'y_bar'}))
```

```python
# 1. How large sample should be taken from a population of 2000 units so that sample estimate
# of population mean (y_bar) differs from its true value by a quantity less than 20 with
# probability 0.95? It is given that an estimate of population variance is 10000.
# (i.e. we have N=2000, apslon(e)=20, 1-alpha=0.95, SY2=10000)
```

```python
N=2000
sy2=10000
e=20
a=0.05/2;a
z=abs(scipy.stats.norm.ppf(a));z
n_WOR=(N*(z**2)*sy2)/((N*e**2)+(z**2*(sy2)))
n_WR=((N-1)*(z**2)*sy2)/(N*e**2)
print("Require no of sample in Without Replacement:\n",n_WOR)
print("Require no of sample in With Replacement:\n",n_WR)
```

```python
N=5000
e=0.1
sy_e=6
z=abs(scipy.stats.norm.ppf(0.025));z
n_WOR=(N*(z**2)*sy_e**2)/((N)+(z**2*(sy_e)**2))
n_WR=((N-1)*(z**2)*(sy_e**2))/(N)
print("Require no of sample in Without Replacement:\n",n_WOR)
print("Require no of sample in With Replacement:\n",n_WR)
```

```python
# Q1). A Loksabha constituency consists of three tehsils with populations sizes 1 = 50000, 2 =
# 35000 and 3 = 25000. A politician Mr. X wants to know the proportion ()of people who
# may vote in favour of him in election. The statistician has taken SRS from three tehsils with
# sample sizes 1 = 2000, 2 = 1500 and 3 = 1500 respectively. Based on opinion survey
# conducted to estimate (), it has been observed that 1050,790 and 450 peoples
# respectively from three tehsils have voted in favour of Mr. X. Using stratified random
# sampling method, estimate () and S.E. of your estimate of (). 
```

```python
N_h=np.array([50000,35000,25000])
n_h=np.array([2000,1500,1500])
w_h=N_h/sum(N_h)
a_h=np.array([1050,790,450])
p_h_hat=a_h/n_h
p_hat_st=sum(w_h*p_h_hat)
print("Estimste of population proportion is",p_hat_st)
b_h=(N_h-n_h)/(N_h*(n_h-1))
var_P_st_hat=sum(w_h**2*b_h*p_h_hat*(1-p_h_hat))
SE_P_st_hat=np.sqrt(var_P_st_hat)
pd.DataFrame({'p_hat_st':[p_hat_st],'SE_P_st_hat':[SE_P_st_hat]})
```

```python
#To find Var(Ybar_st)under Neyman's optimum Allocation
n_h_2=n*W_h*S_h/sum(W_h*S_h);n_h_2
Var_ybar_st_Ney_opt=1/n*(sum(W_h*S_h))**2-1/N*sum(W_h*S_h**2);Var_ybar_st_Ney_opt
n_h=n*N_h*S_h/np.sqrt(C_h)/sum(N_h*S_h/np.sqrt(C_h))
C=6000;c_0=1500
n=(C-c_0)*sum(W_h*S_h/np.sqrt(C_h))/(sum(W_h*S_h*np.sqrt(C_h)));n
n=round(n)
Var_Ybar_st_Ney_opt_2=1/n*(sum(W_h*S_h))**2-1/N*sum(W_h*S_h**2);Var_Ybar_st_Ney_opt_2
print("Var(Ybar_st)under Neyman's optimum Allocation:\n",Var_Ybar_st_Ney_opt_2)
#Relative efficiency 
RE_1=Var_Ybar_SRSWOR/Var_ybar_st_prop;RE_1
RE_2=Var_ybar_st_prop/Var_ybar_st_Ney_opt;RE_2
print("Relative Efficiency Var_ybar_st_prop over Var_Ybar_SRSWOR:\n",RE_1)
print("Relative Efficiency Var_ybar_st_Ney_op over Var_ybar_st_prop:\n",RE_2)
```

```python
# Q3) Table gives the stratification of all farms in a country by farm size and the acreage of corn in
# the sampled farms. Obtain the estimate of average acres of corn per farm in the population
# and its variance. Also, obtain the variances of estimates of population mean under
# proportional allocation and optimum allocation for the same total sample size of 21 farms.
# Hence, obtain the efficiency of given allocation and proportional allocation with respect to
# optimum allocation.

# Farm size           No. of farms()              Acreage of corn in sampled
#  (acres)                                             farms(10 acres)
# 0-40                   12                           3.5, 2.7, 1.1, 1.9
# 41-80                   24                           2.4, 6.9, 3.7, 2.8, 3.4, 2.6
# 81-120                  19                           1, 1.3, 2.9, 1.6, 3.4
# 121-160                 13                           1, 0, 2.3, 0.5
# 161-200                 9                            5.8, 4.9
```

```python
N_h=np.array([12,24,19,13,9])
a1=np.array([3.5,2.7,1.1,1.9])
a2=np.array([2.4,6.9,3.7,2.8,3.4,2.6])
a3=np.array([1,1.3,2.9,1.6,3.4])
a4=np.array([1,0,2.3,0.5])
a5=np.array([5.8,4.9])
n_h=np.array([len(a1),len(a2),len(a3),len(a4),len(a5)])
Y_bar_h=np.array([np.mean(a1),np.mean(a2),np.mean(a3),np.mean(a4),np.mean(a5)])
Sh_2=np.array([np.var(a1),np.var(a2),np.var(a3),np.var(a4),np.var(a5)])
#Est of Population mean
N=sum(N_h)
W_h=N_h/N
Ybar_st=sum(W_h*Y_bar_h)
est_var_Ybar_st=sum((1/N_h-1/N_h)*W_h**2*Sh_2)
est_var_Ybar_st_prop=(1/sum(n_h)-1/N)*sum(W_h)
print("est_var_Ybar_st",est_var_Ybar_st)
print("est_var_Ybar_st_prop",est_var_Ybar_st_prop)
```
